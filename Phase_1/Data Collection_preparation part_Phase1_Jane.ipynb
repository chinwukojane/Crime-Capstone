## Webscraping for Population Census in India as at 2011 from Wikipedia

```python
# import the required libraries
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
# Initialize the WebDriver
driver = webdriver.Chrome()
```

```python
# Open the wikipedia website
driver.get("https://en.wikipedia.org/wiki/2011_census_of_India")
```

```python
# Locate the table with population data
table_element = driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[6]')
```

```python
# Extract data from the table
rows = table_element.find_elements(By.TAG_NAME, 'tr')

data = []

for row in rows:
    # Extract data from each row
    columns = row.find_elements(By.TAG_NAME, 'td')
    row_data = [col.text for col in columns]
    data.append(row_data)
```

```python
# Create a DataFrame
columns = ['Rank','State/UT','Capital','Type','Population', 'Percentage (%)','Males','Females','Sex_Ratio','Literacy_rate', 'Rural population', 'Urban population','Area','Density','Decadal growth rate', ]
df = pd.DataFrame(data, columns=columns)
```

```python
# Display the DataFrame
print(df)
```

```python
#Save the Pandas dataframe in a csv
literacy_rate = 'India_population.csv'
df.to_csv(literacy_rate, index=False)
```

```python
#close the driver
driver.quit()
```

## Webscraping for Literacy rate in India as at 2011 from census2011​ website

```python
# import the required libraries
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
# Initialize the WebDriver
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
# Open the census2011 website
driver.get("https://www.census2011.co.in/literacy.php")
```

```python
# for click on consent button
consent_button=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[1]/div[2]/div[2]/button[1]/p')
consent_button.click()
```

```python
#create empty lists for the features to be scraped from the website
Serial_No = []
State = []
Literacy = []
Male = []
Female = []
Percent_change = []
```

```python
#fetch the serial numbers
sn = driver.find_elements(By.XPATH, '/html/body/div[2]/main/div/table/tbody/tr/td[1]')
for i in sn:
    Serial_No.append(i.text)
    
#fetch the States
state = driver.find_elements(By.XPATH, '/html/body/div[2]/main/div/table/tbody/tr/td[2]')
for i in state:
    State.append(i.text)
    
#fetch the literacy rates
literacy = driver.find_elements(By.XPATH, '/html/body/div[2]/main/div/table/tbody/tr/td[3]')
for i in literacy:
    Literacy.append(i.text)
    
#fetch the literacy rate for male
male = driver.find_elements(By.XPATH, '/html/body/div[2]/main/div/table/tbody/tr/td[4]')
for i in male:
    Male.append(i.text)
    
#fetch the literacy rate for female
female = driver.find_elements(By.XPATH, '/html/body/div[2]/main/div/table/tbody/tr/td[5]')
for i in female:
    Female.append(i.text)
    
#fetch the percent_change column
percent_change = driver.find_elements(By.XPATH, '/html/body/div[2]/main/div/table/tbody/tr/td[6]')
for i in percent_change:
    Percent_change.append(i.text)
```

```python
# Creating Dataframe and display results
df=pd.DataFrame({"Serial No ":Serial_No,"State":State,"Literacy":Literacy,
                "Male":Male,"Female":Female,
                "Percent_change":Percent_change})
df
```

```python
#Save the Pandas dataframe in a csv
literacy_rate = 'Literacy_Rate_India.csv'
df.to_csv(literacy_rate, index=False)
```

```python
#close the driver
driver.quit()
```
